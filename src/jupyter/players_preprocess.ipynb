```python
import pandas as pd

palyers_path_file = "../../data/scout_table.csv"

df = pd.read_csv(palyers_path_file, encoding='utf_16')

# list of columns selected for preprocessing
# columns deselected: 'apelido', 'status_id'
round_col_select = ['rodada_id', 'atleta_id', 'A', 'CA', 'CV', 'DD', 'DP', 'FC', 'FD', 'FF', 'FS', 'FT', 'G', 
                    'GC', 'GS', 'I', 'PE', 'PP', 'RB', 'SG', 'clube_id', 'jogos_num',  'media_num', 
                    'pontos_num', 'preco_num',  'posicao_id',  'variacao_num' ]

df = df[round_col_select]
df.info()
```

```python
# df = df[["rodada_id", "atleta_id", "clube_id", "posicao_id", "pontos_num", "jogos_num", "preco_num"]]

ind = df.set_index(["atleta_id", "rodada_id"])
df_players = pd.DataFrame()
for player in ind.index.get_level_values(0).unique():
    # .copy(deep=True) to supress SettingWithCopyWarning
    df_temp = df[df.atleta_id == player].copy(deep=True) 
    # test to flag the player who plays in a specific round
    df_temp["jogou"] = (df_temp["jogos_num"].diff() != 0) & (df_temp["jogos_num"] != 0)
    df_players = pd.concat([df_players, df_temp], axis=0)
# # discard rows where player didnt play the match
df_scouts = df_players # [df_players.jogou]

```
